```python
import pandas as pd

pd.options.display.max_columns = None
```

```python
nombre = pd.read_csv('../../data/veteranos/nombres_equipos.csv', index_col = 0)
nombre.head(2)
```

```python
caract = pd.read_csv('../../data/veteranos/caracteristicas_jugadores.csv', index_col = 0)
caract.head(2)
```

```python
futuro = pd.read_csv('../../data/veteranos/progresion_jugadores.csv', index_col = 0)
futuro.head(2)
```

```python
precio = pd.read_csv('../../data/veteranos/precio_jugadores.csv', index_col = 0)
precio.head(2)
```

```python
valorac = pd.read_csv('../../data/veteranos/valoraciones_jugadores.csv', index_col = 0)
precio.head(2)
```

```python
veteranos = pd.concat([nombre, caract, futuro, precio, valorac], axis=1)
veteranos
```

```python
veteranos.to_csv('../../data/veteranos/jugadoresveteranos.csv')
```
