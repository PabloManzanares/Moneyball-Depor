```python
import pandas as pd

pd.options.display.max_columns = None
```

```python
nombre = pd.read_csv('../../data/promesas/nombres_equipos.csv', index_col = 0)
nombre.head(2)
```

```python
caract = pd.read_csv('../../data/promesas/caracteristicas_jugadores.csv', index_col = 0)
caract.head(2)
```

```python
futuro = pd.read_csv('../../data/promesas/progresion_jugadores.csv', index_col = 0)
futuro.head(2)
```

```python
precio = pd.read_csv('../../data/promesas/precio_jugadores.csv', index_col = 0)
precio.head(2)
```

```python
valorac = pd.read_csv('../../data/promesas/valoraciones_jugadores.csv', index_col = 0)
valorac.head(2)
```

```python
jovenespromesas = pd.concat([nombre, caract, futuro, precio, valorac], axis=1)
jovenespromesas
```

```python
jovenespromesas.to_csv('../../data/promesas/jovenespromesas.csv')
```
