```python
import pandas as pd

pd.options.display.max_columns = None
```

```python
nombre = pd.read_csv('../../data/precio/nombres_equipos.csv', index_col = 0)
nombre.head(2)
```

```python
paisedad = pd.read_csv('../../data/precio/paisedad_jugadores.csv', index_col = 0)
paisedad.head(2)
```

```python
futuro = pd.read_csv('../../data/precio/progresion_jugadores.csv', index_col = 0)
futuro.head(2)
```

```python
precio = pd.read_csv('../../data/precio/precios_jugadores.csv', index_col = 0)
precio.head(2)
```

```python
valorac = pd.read_csv('../../data/precio/valoraciones_jugadores.csv', index_col = 0)
valorac
```

```python
jovenes = pd.concat([nombre, paisedad, futuro, precio, valorac], axis=1)
jovenes.head()
```

```python
jovenes.to_csv('../../data/precio/jovenes.csv')
```
