```python
import pandas as pd
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler

pd.options.display.max_columns = None
```

```python
df = pd.read_csv("../../data/precio/dfproc.csv", index_col = 0)
df.head()
```

```python
df = df.drop(['Nombre jugador', 'Equipo actual', 'pais'], axis =1)
df.head()
```

```python
scaler = StandardScaler()
```

```python
df.columns
```

```python
scaler.fit(df[['edad', 'Progresión', 'Valoración actual', 'Ataque', 'Regate',
       'Aceleración', 'Potencia tiro', 'Agresividad', 'Defensa']])
```

```python
X_escaladas = scaler.transform(df[['edad','Progresión', 'Valoración actual', 'Ataque', 'Regate',
       'Aceleración', 'Potencia tiro', 'Agresividad', 'Defensa']])
X_escaladas
```

```python
df[['edad', 'Progresión', 'Valoración actual', 'Ataque', 'Regate',
       'Aceleración', 'Potencia tiro', 'Agresividad', 'Defensa']] = X_escaladas
df.head()
```

```python
df.to_csv('../../data/precio/dfnorm.csv')
```
